```python
import pandas as pd
```

```python
path = "/content/CSV_BR.txt"
df = pd.read_csv(path)
```

```python
df.head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['dt_hr_relato'] = pd.to_datetime(df['dt_hr_relato'], format='%d/%m/%Y')
```

```python
df['year'] = df['dt_hr_relato'].dt.year
plt.figure(figsize=(10, 6))
sns.countplot(x='year', data=df, palette='viridis')
plt.title('Distribuição de Relatos por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Relatos')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='boolean_praia_suja', data=df, palette='viridis')
plt.title('Quantidade de Relatos de Praias Sujas vs. Limpos')
plt.xlabel('Praia Suja (1) vs. Limpa (0)')
plt.ylabel('Quantidade de Relatos')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='boolean_envolve_animais', data=df, palette='viridis')
plt.title('Quantidade de Relatos Envolvendo Animais')
plt.xlabel('Envolve Animais (1) vs. Não Envolve (0)')
plt.ylabel('Quantidade de Relatos')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['nr_longitude'], df['nr_latitude'], c='blue', alpha=0.5)
plt.title('Distribuição Geográfica dos Relatos')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='boolean_praia_suja', y='nr_likes', data=df, palette='viridis')
plt.title('Número de Likes por Tipo de Relato')
plt.xlabel('Praia Suja (1) vs. Limpa (0)')
plt.ylabel('Número de Likes')
plt.show()
```

```python
df['year'] = df['dt_hr_relato'].dt.year
plt.figure(figsize=(10, 6))
sns.boxplot(x='year', y='nr_likes', data=df, palette='viridis')
plt.title('Distribuição de Likes por Ano')
plt.xlabel('Ano')
plt.ylabel('Número de Likes')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='nr_longitude', y='nr_latitude', hue='boolean_praia_suja', data=df, palette='viridis')
plt.title('Correlação entre Latitude e Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Praia Suja')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='id_usuario', data=df, palette='viridis', order=df['id_usuario'].value_counts().index)
plt.title('Quantidade de Relatos por Usuário')
plt.xlabel('ID do Usuário')
plt.ylabel('Quantidade de Relatos')
plt.xticks(rotation=90)
plt.show()
```

```python
df['month'] = df['dt_hr_relato'].dt.month
plt.figure(figsize=(10, 6))
sns.countplot(x='month', data=df, palette='viridis')
plt.title('Distribuição de Relatos por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Relatos')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='boolean_praia_suja', hue='boolean_envolve_animais', data=df, palette='viridis')
plt.title('Número de Relatos por Tipo de Relato (Praia Suja e Envolve Animais)')
plt.xlabel('Praia Suja (1) vs. Limpa (0)')
plt.ylabel('Quantidade de Relatos')
plt.legend(title='Envolve Animais')
plt.show()
```

```python
df = df.drop(columns=['ds_relato', 'blob_foto', 'id_usuario', 'id_relato'])
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlação')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='nr_longitude', y='nr_latitude', hue='boolean_praia_suja', palette='coolwarm', s=100)
plt.title('Localização dos Relatos - Praia Suja')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Praia Suja', loc='upper right', labels=['Não', 'Sim'])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='nr_longitude', y='nr_latitude', hue='boolean_envolve_animais', palette='viridis', s=100)
plt.title('Localização dos Relatos - Envolve Animais')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Envolve Animais', loc='upper right', labels=['Não', 'Sim'])
plt.show()
```
