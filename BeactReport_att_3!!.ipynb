```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd

def plot_map_markers(data):
    latitude = data['nr_latitude'].tolist()
    longitude = data['nr_longitude'].tolist()
    descricao = data['ds_relato'].tolist()
    praia_suja = data['boolean_praia_suja'].tolist()
    envolve_animais = data['boolean_envolve_animais'].tolist()

    start_coords = [latitude[0], longitude[0]]

    m = folium.Map(location=start_coords, zoom_start=12)

    marker_cluster = MarkerCluster().add_to(m)

    for lat, lon, desc, suja, animais in zip(latitude, longitude, descricao, praia_suja, envolve_animais):
        if suja and animais:
            color = 'purple'  # Ambos
        elif suja:
            color = 'red'  # Praia suja
        elif animais:
            color = 'blue'  # Envolve animais
        else:
            color = 'green'

        folium.Marker(
            location=[lat, lon],
            popup=desc,
            icon=folium.Icon(color=color)
        ).add_to(marker_cluster)

    return m

map_with_markers = plot_map_markers(df)

map_with_markers.save("map_with_markers.html")

map_with_markers
```

# **MODELO DE PREDIÇÃO**

```python
df_cleaned['latitude_interval'] = pd.cut(df_cleaned['nr_latitude'], bins=10)
df_cleaned['longitude_interval'] = pd.cut(df_cleaned['nr_longitude'], bins=10)
```

```python
lixo_counts = df_cleaned[df_cleaned['boolean_praia_suja'] == 1].groupby(['latitude_interval', 'longitude_interval']).size().unstack(fill_value=0)
```

```python
animais_counts = df_cleaned[df_cleaned['boolean_envolve_animais'] == 1].groupby(['latitude_interval', 'longitude_interval']).size().unstack(fill_value=0)
```

```python
X = df_cleaned[['nr_latitude', 'nr_longitude', 'nr_likes']]
y_lixo = df_cleaned['boolean_praia_suja']
y_animais = df_cleaned['boolean_envolve_animais']
```

```python
X_train_lixo, X_test_lixo, y_train_lixo, y_test_lixo = train_test_split(X, y_lixo, test_size=0.3, random_state=42)
X_train_animais, X_test_animais, y_train_animais, y_test_animais = train_test_split(X, y_animais, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_lixo = scaler.fit_transform(X_train_lixo)
X_test_lixo = scaler.transform(X_test_lixo)
X_train_animais = scaler.fit_transform(X_train_animais)
X_test_animais = scaler.transform(X_test_animais)
```

```python
def train_and_evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Relatório de Classificação para {model_name}:")
    print(classification_report(y_test, y_pred))
    print(f"Matriz de Confusão para {model_name}:")
    print(confusion_matrix(y_test, y_pred))
    print(f"Acurácia para {model_name}: {accuracy_score(y_test, y_pred)}\n")
```

```python
models = {
    "K-Nearest Neighbors": KNeighborsClassifier(),
}
```

```python
print("Previsão de Lixo:\n")
for model_name, model in models.items():
    train_and_evaluate_model(model, X_train_lixo, X_test_lixo, y_train_lixo, y_test_lixo, model_name)
```

```python
print("Previsão de Ocorrências com Animais:\n")
for model_name, model in models.items():
    train_and_evaluate_model(model, X_train_animais, X_test_animais, y_train_animais, y_test_animais, model_name)
```
