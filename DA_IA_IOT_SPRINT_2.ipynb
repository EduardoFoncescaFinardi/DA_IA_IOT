```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

features = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']
x = df_sub.loc[:, features].values
x = StandardScaler().fit_transform(x)

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf_sub = pd.DataFrame(data=principalComponents, columns=['principal component 1', 'principal component 2'])

plt.figure(figsize=(8,6))
plt.scatter(principalDf_sub['principal component 1'], principalDf_sub['principal component 2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
df_sub_1 = pd.get_dummies(df_sub, columns=['Education', 'Marital_Status'])

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_sub_1[features] = scaler.fit_transform(df_sub_1[features])
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(df_sub_1[features])
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Mtodo do Cotovelo')
plt.xlabel('Nmero de clusters')
plt.ylabel('WCSS')
plt.show()

kmeans = KMeans(n_clusters=3, random_state=0).fit(df[features])
df['Cluster'] = kmeans.labels_
```

```python
n_clusters = 3

kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
cluster_labels = kmeans.fit_predict(x)
```

```python
df_sub_1['Cluster'] = cluster_labels
```

```python
cluster_analysis = df_sub_1.groupby('Cluster').mean()
cluster_analysis
```

```python
# Supondo que voc tenha os componentes principais em principalDf_sub como antes
principalDf_sub['Cluster'] = cluster_labels

# Visualizao
plt.figure(figsize=(10, 8))
plt.scatter(principalDf_sub['principal component 1'], principalDf_sub['principal component 2'], c=principalDf_sub['Cluster'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Visualizao dos Clusters')
plt.colorbar(label='Cluster')
plt.show()
```
